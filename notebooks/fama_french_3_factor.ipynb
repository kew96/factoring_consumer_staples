# Fama-French Three-Factor Model with Markowitz Optimization

```python
import pandas as pd
import matplotlib.pyplot as plt

import sys; sys.path.insert(0, '..')
from src.features import fama_french
```

```python
tfm = fama_french.ThreeFactorModel()
```

```python
rets = tfm.max_sharpe_portfolios(2005, 2020, return_sharpe=True)
```

```python
mean = rets.actual_ret.mean()
std = rets.actual_ret.std()
print(f'Actual:\t\t{mean:.4f} ({std:.4f}) ({mean-1.96*std:.4f}, {mean+1.96*std:.4f})')

mean = rets.expected_ret.mean()
std = rets.expected_ret.std()
print(f'Expected:\t{mean:.4f} ({std:.4f}) ({mean-1.96*std:.4f}, {mean+1.96*std:.4f})')
```

```python
rets.sort_index().plot(y=['actual_ret', 'expected_ret'], figsize=(15, 10), grid=True, legend=True)
```

```python
rets.corr()
```

```python
rets.sort_index().sharpe.plot(figsize=(15, 10), grid=True, legend=False)
```
