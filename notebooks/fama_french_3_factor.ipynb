```python
import pandas as pd
import matplotlib.pyplot as plt

import sys; sys.path.insert(0, '..')
from src.features import fama_french
```

```python
tfm = fama_french.ThreeFactorModel()
```

```python
tfm.p_values.sort_index(ascending=False).plot(kind='barh', figsize=(15, 100), grid=True)
```

```python
rets = tfm.max_sharpe_portfolios(2005, 2012)
```

```python
mean = rets.ret.mean()
std = rets.ret.std()
print(f'{mean} ({std}): \t ({mean-1.96*std}, {mean+1.96*std})')
```

```python
rets.sort_index().plot(figsize=(15, 10), grid=True, legend=False)
```
